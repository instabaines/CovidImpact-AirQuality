```python
def get_tweets(screen_names):
    a_ids = []
    a_times = []
    a_texts = []
    a_favs = []
    a_retweets = []
    a_entities = []
    a_languages = []
    
    for name in screen_names:
        
        ids = []
        times = []
        texts = []
        favs = []
        retweets = []
        entities = []
        langs = []
        
        tweets = tweepy.Cursor(api.user_timeline, screen_name = name, since = '2020-01-01', wait_on_rate_limit=True).items(50000)
        for tweet in tweets:
            id_ = tweet.id_str
            time = tweet.created_at
            text_ = tweet.text
            fav = tweet.favorite_count
            retweet = tweet.retweet_count
            entities_ = tweet.entities
            lang = tweet.lang
            
            ids.append(id_)
            times.append(time)
            texts.append(text_)
            favs.append(fav)
            retweets.append(retweet)
            entities.append(entities_)
            langs.append(lang)
            
        a_ids.append(ids)
        a_times.append(times)
        a_texts.append(texts)
        a_favs.append(favs)
        a_retweets.append(retweets)
        a_entities.append(entities)
        a_languages.append(langs)

    a_ids = [y for x in a_ids for y in x]
    a_times = [y for x in a_times for y in x]
    a_texts = [y for x in a_texts for y in x]
    a_favs = [y for x in a_favs for y in x]
    a_retweets = [y for x in a_retweets for y in x]
    a_entities = [y for x in a_entities for y in x]
    a_languages = [y for x in a_languages for y in x]
        
    df = pd.DataFrame({'id':a_ids, 'time':a_times, 'text':a_texts, 'favourite_count':a_favs,
                       'retweet_count':a_retweets,'entities':a_entities, 'lang':a_languages})
    return df
```
